```python
#4 Найдите количество покемонов чья скорость больше 80
len(df[df['Speed']>80])
```

```python
#5 Создайте новый столбец Total_Stats, который будет представлять собой сумму 
#HP, Attack, Defense и Speed для каждого покемона. Выведите названия покемонов у которых Total_Stats больше 400.
df['Total_Stats']=df['HP'] + df['Attack'] + df['Defense'] + df['Speed']
df[df['Total_Stats'] > 400]['Name']
```

```python
#6 На каком индексе находится Пикачу (Pikachu) 
df[df['Name'].str.contains('Pikachu', case=False)].index[0]
```

```python
#Дополнительные задания:
#7. Создайте столбец Is_Fast, который будет содержать значение True, если скорость покемона больше 80, иначе False.
df['Is_Fast']=df['Speed']>80
```

```python
# 8. Вычислите средний показатель атаки среди "быстрых" покемонов у которых скорость больше 80.
df[df['Is_Fast']==True][['Attack']].mean()
```

```python
# 9. Сравните максимальное и минимальное значения speed между покемонами типа "Water" и "Fire". Какой тип в среднем быстрее?

# Сравнение максимального и минимального значения speed между покеменом типа "Water" и "Fire"
df[df['Type 1'].isin(['Water'])]['Speed'].max()==df[df['Type 1'].isin(['Fire'])]['Speed'].max()
# Какой тип в среднем быстрее?
mean_of_water=df[df['Type 1'].isin(['Water'])]['Speed'].mean()
mean_of_fire=df[df['Type 1'].isin(['Fire'])]['Speed'].mean()
if(mean_of_fire>mean_of_water):
    print('Fire is faster')
elif(mean_of_fire<mean_of_water):
    print('Water is faster')
else:
    print('Print they are equal')
print(mean_of_fire)
```
